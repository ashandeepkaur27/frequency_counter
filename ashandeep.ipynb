```python
print("------------ Frequency Counter ------------")

text = input("Enter any sentence or paragraph: ")

words = text.split()

frequency = {}

for word in words:
    word = word.lower()
    if word in frequency:
        frequency[word] = frequency[word] + 1
    else:
        frequency[word] = 1

print("\n----------- Result -----------")

for key in frequency:
    print(key, ":", frequency[key])

print("\nTotal unique words:", len(frequency))

all_words = list(frequency.keys())

print("\nWords found:")
for w in all_words:
    print("-", w)

print("\nMost frequent word:")

max_count = 0
max_word = ""

for w in frequency:
    if frequency[w] > max_count:
        max_count = frequency[w]
        max_word = w

print(max_word, "appears", max_count, "times")

print("\nSorted frequency:")

sorted_items = sorted(frequency.items())

for item in sorted_items:
    print(item[0], ":", item[1])

print("\nProgram Completed.")
```
